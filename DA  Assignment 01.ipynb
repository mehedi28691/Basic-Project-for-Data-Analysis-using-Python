```python
#১.ভ্যারিয়েবল ও ডাটা টাইপ: আপনার নাম, বয়স, এবং ছাত্র কিনা (True/False) — 
#এই তথ্যগুলো ভ্যারিয়েবল হিসেবে সংরক্ষণ করুন। প্রতিটি ভ্যারিয়েবলের টাইপ print(type(...)) দিয়ে দেখান।
name = "mehedi"
age = 25
is_student = True

print(type(name))  
print(type(age))    
print(type(is_student))
```

```python
#২.অ্যারিথমেটিক অপারেশন: আপনার বয়স দিয়ে একটি সংখ্যা গুণ, যোগ, ভাগ, বিয়োগ করে প্রিন্ট করুন।
age = 25
print ("multiplication:", age * 5)
print ("addition:", age + 5)
print ("division:", age / 5)
print ("subtraction:", age - 5)
```

```python
#৩.কম্পারিজন অপারেটর: বয়স ১৮ এর বেশি কিনা তা চেক করুন এবং ফলাফল প্রিন্ট করুন। ==, !=, <, > অপারেটর ব্যবহার করে ২/৩টি তুলনা দেখান।
age = 25
print (age == 18)
print (age <= 18)
print (age >= 18)
```

```python
#৪.লজিক্যাল অপারেটর: দুটি কন্ডিশন তৈরি করুন এবং and, or, not দিয়ে কিভাবে কাজ করে তা দেখান।
age = 25
print (age <= 18 and age >= 18 )
print (age == 18 or age <= 18 )
print (age != 18 or age >= 18 )
```

```python
#৫.অ্যাসাইনমেন্ট অপারেটর: একটি ভ্যারিয়েবল নিন এবং +=, -=, *=, /= দিয়ে মান পরিবর্তন করে দেখান।
x = 25
x += 5 #30
x -= 10 #20
x *= 10 #200
x /= 10 #20

print (int(x))
```

```python
#৬.আইডেন্টিটি অপারেটর: দুটি ভিন্ন ভ্যালু যুক্ত ভ্যারিয়েবল তৈরি করে is এবং is not অপারেটর দিয়ে দেখান।
x = [2, 4, 6]
y = [1, 3, 6]
z = x

print (x is y)
print (z is x)
print (x is not y)
```

```python
#৭.মেম্বারশিপ অপারেটর: একটি লিস্ট তৈরি করুন এবং কোন একটি উপাদান সেই লিস্টে আছে কিনা তা in, not in দিয়ে দেখান।
x = [2, 4, 6, 8, 10]

print (2 in x)
print (5 not in x)
```
